# Computer Vision Problem - The Edge Of Laddoos

###  The python code is attached along with this file

##### Step 1: Load the libraries

```python
import cv2
from PIL import Image
```

##### Step 2: Get  the image paths and setup output directory

```python
input = 'laddu.jpeg'
output = 'output_laddu.jpeg'
optimized = 'optimized_laddu.jpeg'
```

##### Step 3: Display  the basic input file

```python
 Image.open(input)
```

##### Step 4: Read  the image_ input file

```python
image = cv2.imread(input)
```

##### Step 5: Create Edges and draw the contour of  the image_ input file

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (11, 11), 0)
canny = cv2.Canny(blurred, 30, 150)
contours, hierarchy = cv2.findContours(canny.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
```

##### Step 6: Display the output Image file file

```python
cv2.imwrite(output, image)
Image.open(output)
```

##### Step 7: Optimize the edges

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (11, 11), 0)
canny = cv2.Canny(blurred, 30, 90)
contours, hierarchy = cv2.findContours(canny.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
image = cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
```

##### Step 8: Display the output optimized Image file

```python
cv2.imwrite(optimized, image)
Image.open(optimized)
```
